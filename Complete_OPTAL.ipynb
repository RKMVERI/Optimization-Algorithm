## **ARMIJO'S ALGORITHM OF LINE SEARCH**  code and example showing optimizing the step-length with the following two functions:- $$ x^2$$ and $$x^4 + x^2 $$___Mahendra Nandi

```python
f = lambda x : x**2
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2
def f1(x):
    return 2*x

# armijo gives a optimal steplength at a given point in a given direction
#            [in case of univariate we use max. decent direction,i.e, the gradient descent] 
#            it took the function,rho,and an initially guessed lambda as input and returns 
#            optimal lambda for that situation

def armijo(f,f1,rho,bita,x):
    s=-f1(x)
    lamda=bita
    l=[]
    while ( f(x+lamda*s)-f(x) > lamda*rho*f1(x)*s ):
        l.append(lamda)
        lamda=lamda/2
    return l , lamda   
        
    
# plots   

list_of_lamda,optimal_lamda=armijo(f,f1,0.02,100,4)

plt.plot(list_of_lamda),plt.xlabel("iteration no."),plt.ylabel("value of step length"),plt.title("Showing graph of step sizes chosen with iteration in \nArmijo’s line search for function-1"),plt.show()






#### checking with another function



def fn(x):
    return x**4+x**2
def fn1(x):
    return 4*(x**3)+2*x


list_of_lamda2,optimal_lamda2=armijo(fn,fn1,0.02,100,4)

plt.plot(list_of_lamda2),plt.xlabel("iteration no."),plt.ylabel("value of step length"),plt.title("Showing graph of step sizes chosen with iteration in \nArmijo’s line search for function-2"),plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**4
def f1(x):
    return 4*(x**3)

def f2(x):
    return 12*(x**2)

#############################################################################


def GD_with_CD_and_fixed_lamda(x,f,f1,f2,bita,epsilon):
    l1=[]
    s=-f1(x)
    lamda=bita
    i=0
    #bita=(f2(x)**2)/(f1(x)**2)
    while( np.abs(f(x+bita*s)-f(x) )>epsilon ):
        i=i+1
        l1.append(x)
        ff1=(f1(x)**2)
        x=x+lamda*s
        ff2=(f1(x)**2)
        bita=ff2/ff1
        s=-f1(x)+bita*s
    return   [plt.figure(figsize=(12,3)),plt.subplot(121),plt.plot(l1,"g"), plt.subplot(122),
            plt.plot(  l1,[f(m) for m in l1]  , 
                     np.linspace(-l1[0],l1[0],40),[f(m) for m in np.linspace(-l1[0],
                                l1[0],40)] ,(".r")),plt.show() ],x,i

GD_with_CD_and_fixed_lamda(2,f,f1,f2,0.1,0.0001)   
```

```python
def ARMIJO(F,G,rho,bita,X):
    S=-G(X)
    lamda=bita
    l=[]
    while ( F(X+lamda*S)-F(X) > lamda*rho*(G(X).T @ S) ):
        l.append(lamda)
        lamda=lamda/2
        #print(lamda)
    return l[-1]
```
